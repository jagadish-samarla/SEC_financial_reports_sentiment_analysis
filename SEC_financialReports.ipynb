Used word tokenizer and sentence tokeniser to calculate required features of the data

```python
from nltk import word_tokenize

def clean_and_tokenize(text):
  text = "".join([word for word in text if word not in string.punctuation])
  token = word_tokenize(text)
  token = [word for word in token if word not in stopwords and len(word)>1]
  return token

df['report_no_stopwords'] = df['reports'].apply(lambda x: clean_and_tokenize(x.lower()))
df.head()
```

```python
positive_np = master_dict['Word'][master_dict.Positive>0].tolist()
#positive_np = positive_dict.to_numpy()
positive_np = [word.lower() for word in positive_np]
positive_np
```

```python
negative_np = master_dict['Word'][master_dict.Negative>0].tolist()
#negative_np = negative_dict.to_numpy().tolist()
negative_np = np.array(negative_np)
negative_np = [word.lower() for word in negative_np]
negative_np
```

```python

def positive_score(tokens):
    count = 0

    for word in tokens:
      if word in positive_np:
        count+=1
    return count

df['positive_score'] = df['report_no_stopwords'].apply(lambda x: positive_score(x))

df.head()
```

```python

def negative_score(tokens):
    count = 0
    for word in tokens:
      if word in negative_np:
        count+=1
    return count

df['negative_score'] = df['report_no_stopwords'].apply(lambda x: negative_score(x))

df.head()
```

```python
for index, row in df.iterrows():
  df.loc[index, 'polarity_score'] = (row['positive_score']-row['negative_score'])/((row['positive_score']+row['negative_score'])+0.000001)

```

```python
from nltk import sent_tokenize, word_tokenize
#sample_text = 'this is to check. sentence tokeniser. using nltk'
def word_count(text):
  tokens = word_tokenize(text)
  tokens = [tokens for word in tokens if word not in string.punctuation]
  return len(tokens)
```

```python
df['word_count'] = df['reports'].apply(lambda x: word_count(x))
```

```python
df.head()
```

```python
def sentence_count(text):
  tokens = sent_tokenize(text)
  tokens = [tokens for sentences in tokens if sentences not in string.punctuation]
  return len(tokens)
```

```python
df['sentence_count'] = df['reports'].apply(lambda x: sentence_count(x))
```

```python
df.head()
```

```python
for index, row in df.iterrows():
  df.loc[index, 'average_sentence_length'] = (row['word_count']/row['sentence_count'])
```

```python
df.head()
```

```python
def syllable(text):
  """This function counts the number of syllables for each word in the text"""
  syllable_list =[]
  #text = "".join([word for word in text if word not in string.punctuation])
  #tokens = re.split('\W+', text)
  #tokens = [tokens.lower() for tokens in text]
  tokens = text.split()
  for word in tokens:
    syllable_count = 0
    vowels = "aeiouy"
    if word[0] in vowels:
        syllable_count += 1
    for index in range(1, len(word)):
        if word[index] in vowels and word[index - 1] not in vowels:
            syllable_count += 1
    if word.endswith("e"):
       syllable_count -= 1
    if syllable_count == 0:
        syllable_count += 1
    syllable_list.append(syllable_count)
  return syllable_list


def add_complex(count_list):
  """this function only adds number of complex words in the given list"""
  complex_count = 0
  for i in count_list:
    if i > 2:
      complex_count += 1
  return complex_count


#print(add_complex(syllable('similarity polarity redemption check output')))
```

```python
df['complex_count'] = df['reports'].apply(lambda x:add_complex(syllable(x)))
```

```python
#this is calculated as per given formula in Text Analysis
for index, row in df.iterrows():
  df.loc[index, 'percentage_of_complex_words'] = ((row['complex_count']/row['word_count']))

df.head()
```

```python
for index, rows in df.iterrows():
  df.loc[index, 'fog_index'] = (rows['average_sentence_length']+rows['percentage_of_complex_words'])*0.4
df.head()
```

```python
uncertain = pd.read_excel('/content/drive/MyDrive/BlackCoffer-assignment/uncertainty_dictionary.xlsx')
uncertain.tail()
```

```python
uncertain_np = uncertain['Word'].tolist()
#negative_np = negative_dict.to_numpy().tolist()
uncertain_np = np.array(uncertain_np)
uncertain_np = [word.lower() for word in uncertain_np]

def uncertainty_score(tokens):
    count = 0

    for word in tokens:
      if word in uncertain_np:
        count +=1
        
    return count

df['uncertainty_score'] = df['report_no_stopwords'].apply(lambda x: uncertainty_score(x))

```

```python
constraining = pd.read_excel('/content/drive/MyDrive/BlackCoffer-assignment/constraining_dictionary.xlsx')
constraining_np = constraining['Word'].tolist()
#negative_np = negative_dict.to_numpy().tolist()
constraining_np = np.array(constraining_np)
constraining_np = [word.lower() for word in constraining_np]

def constraining_score(tokens):
    count = 0

    for word in tokens:
      if word in constraining_np:
        count +=1
        
    return count

df['constraining_score'] = df['report_no_stopwords'].apply(lambda x: constraining_score(x))
df.head()
```

```python
for index, rows in df.iterrows():
  df.loc[index, 'positive_word_proportion'] = rows['positive_score']/rows['word_count']

#df.head()
```

```python
for index, rows in df.iterrows():
  df.loc[index, 'uncertainty_word_proportion'] = rows['uncertainty_score']/rows['word_count']

#df.head()
```

```python
for index, rows in df.iterrows():
  df.loc[index, 'constraining_word_proportion'] = rows['constraining_score']/rows['word_count']

df.head()
```

```python

def constraining_list(tokens):
    constraining_list = []

    for word in tokens:
      if word in constraining_np:
        constraining_list.append(word)

    return constraining_list

df['constraining_words_whole_report'] = df['report_no_stopwords'].apply(lambda x: constraining_list(x))
df.head()
```
